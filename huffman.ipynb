```python
#huffman to get the entropy of a 3:1 biased coin

import math

def create_nodes(n):
    l = []
    for k in range(n+1):
        i = math.comb(n,k)
        l += [(3**k,(),())]*i
    l.sort()
    return l

def huff_one_step(l):
    a = l.pop(0)
    b = l.pop(0)
    c = (a[0]+b[0],a,b)
    l.insert(0,c)
    l.sort()
    return l

def print_indented_aux(l,i,t,s):
    if l==():
        return 0
    #for j in range(i):
        #print('\t',end='')
    if l[1]!=():
        print(f'{t}+---{i}:{l[0]}')
    else:
        print(f'{t}----{i}:{l[0]}')
    e = 0
    e += print_indented_aux(l[1],i+1,s+'+--',s+'|  ')
    e += print_indented_aux(l[2],i+1,s+'+--',s+'   ')
    if e==0:
        e = i*l[0]
    return e

def print_indented(l):
    e = print_indented_aux(l,0,'+--','   ')
    return e

def huffman_3of4(n):
    l = create_nodes(n)
    while len(l)>1:
        l = huff_one_step(l)
    # l is a list of nodes, now just one node. get the node
    l = l[0]
    e = print_indented(l)
    return e/4**n/n

print(f'predicted: {-((1/4)*math.log2(1/4)+(3/4)*math.log2(3/4)):.6f}')

#for n in range(4,8):
#    e = huffman_3of4(n)
#    print(f'measured:  {e:.6f}')

e = huffman_3of4(8)
print(f'measured:  {e:.6f}')

```

```python
# Chloe and Drew game

def score_chloe_drew(d_coin,c_coin,d_choice,c_choice):
    correct = 0
    if (d_coin==c_choice):
        correct += 1
    if (c_coin==d_choice):
        correct += 1
    return (correct==1)

def strategy_chloe_drew(d_coin,c_coin):
    d_choice = d_coin
    c_choice = (1-c_coin)
    return (d_choice,c_choice)

def ennumerate_chloe_drew():
    correct = 0
    for c_coin in range(2):
        for d_coin in range(2):
            choices = strategy_chloe_drew(d_coin,c_coin)
            if score_chloe_drew(d_coin,c_coin,choices[0],choices[1]):
                correct += 1
    return correct

print(ennumerate_chloe_drew())

            
```
